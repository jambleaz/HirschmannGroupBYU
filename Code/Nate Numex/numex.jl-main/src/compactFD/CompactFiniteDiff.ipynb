```julia
function d4(n, h)
    dl = fill(0.25,n-1)
    du = fill(0.25,n-1)
    d = ones(n)
    dl[n-1] = 3.0
    du[1] = 3.0
    A = Tridiagonal(dl, d, du)
    
    B = zeros(n,n)
    hm1 = 1.0/h
    t1 = 0.75*hm1
    for i = 2:n-1
        B[i-1,i] = t1
        B[i+1,i] = -t1
    end
    
    B[1,1] = -17.0/6.0*hm1
    B[2,1] = -t1
    B[1,2] = 1.5*hm1
    B[1,3] = 1.5*hm1
    B[1,4] = -1.0/6.0*hm1

    B[n-1,n] = t1
    B[n,n-3] = 1.0/6.0*hm1
    B[n,n-2] = -1.5*hm1
    B[n,n-1] = -1.5*hm1
    B[n,n] = 17.0/6.0*hm1
    
    return A, B
end
```

```julia
function d6(n, h)
    third = 1.0/3.0
    dl = fill(third, n-1)
    du = fill(third, n-1)
    d = ones(n)
    dl[1] = 2.0/11.0
    dl[n-2] = 2.0/11.0
    dl[n-1] = 5.0
    du[1] = 5.0
    du[2] = 2.0/11.0
    du[n-1] = 2.0/11.0
    A = Tridiagonal(dl, d, du)
    
    hm1 = 1.0/h
    B = zeros(n,n)
    t1 = 1.0/36.0*hm1
    t2 = 7.0/9.0*hm1
    for j = 3:n-2
        B[j-2,j] =  t1
        B[j-1,j] =  t2
        B[j+1,j] = -t2
        B[j+2,j] = -t1
    end
    
    B[1,1] = -197.0/60.0*hm1
    B[2,1] = -20.0/33.0*hm1
    B[3,1] = -t1
    
    B[1,2] = -5.0/12.0*hm1
    B[2,2] = -35.0/132.0*hm1
    B[3,2] = -t2
    B[4,2] = -t1
    
    B[1,3] = 5.0*hm1
    B[2,3] = 34.0/33.0*hm1
    
    B[1,4] = -5.0/3.0*hm1
    B[2,4] = -7.0/33.0*hm1
    
    B[1,5] = 5.0/12.0*hm1
    B[2,5] = 2.0/33.0*hm1
    
    B[1,6] = -1.0/20.0*hm1
    B[2,6] = -1.0/132.0*hm1
    
    B[n-1,n-5] = 1.0/132.0*hm1
    B[n,n-5] = 1.0/20.0*hm1
    
    B[n-1,n-4] = -2.0/33.0*hm1
    B[n,n-4] = -5.0/12.0*hm1
    
    B[n-1,n-3] = 7.0/33.0*hm1
    B[n,n-3] = 5.0/3.0*hm1
    
    B[n-1,n-2] = -34.0/33.0*hm1
    B[n,n-2] = -5.0*hm1
    
    B[n-3,n-1] = t1
    B[n-2,n-1] = t2
    B[n-1,n-1] = 35.0/132.0*hm1
    B[n,n-1] = 5.0/12.0*hm1
    
    B[n-2,n] = t1
    B[n-1,n] = 20.0/33.0*hm1
    B[n,n] = 197.0/60.0*hm1
    
    return A, B
end  
    
```

```julia
function dd4(n,h)
    dl = fill(0.1,n-1)
    du = fill(0.1,n-1)
    d = ones(n)
    dl[n-1] = 10.0
    du[1] = 10.0
    A = Tridiagonal(dl, d, du)
    
    B = zeros(n,n)
    hm2 = 1.0/(h*h)
    a = 6.0/5.0
    for i = 2:n-1
        B[i,i-1] = a*hm2
        B[i,i] = -2.0*a*hm2
        B[i,i+1] = a*hm2
    end
    
    # points 1, N
    B[1,1] = 145.0/12*hm2
    B[1,2] = - 76.0/3*hm2
    B[1,3] = 29.0/2*hm2
    B[1,4] = -4.0/3*hm2
    B[1,5] = 1.0/12*hm2
    
    B[n,n] = 145.0/12*hm2
    B[n,n-1] = - 76.0/3*hm2
    B[n,n-2] = 29.0/2*hm2
    B[n,n-3] = -4.0/3*hm2
    B[n,n-4] = 1.0/12*hm2
    
    return A,B
end
```

```julia
function ftest(x)
    f = similar(x)
    w = 2.0*pi
    @. f = sin(w*x)
    df = similar(x)
    @. df = w*cos(w*x)
    return f, df
end
```

```julia
function fpow(x, p)
    a = pi/6.0
    f = similar(x)
    df = similar(x)
    @. f = (x-a)^p
    @. df = p*(x-a)^(p-1)
    return f, df
end
```

```julia
using Random
function addnoise(f, amp)
    
    g = similar(f)
    
    rng = MersenneTwister(1234)
    b = rand!(rng, zeros(length(f)))
    @. g = f + amp * (b-0.5)
    
    
    # @. g = f + amp*sin(10.0*x1)*cos(25.0*x1) + amp*cos(13.0*x1)
    return g
end
```
